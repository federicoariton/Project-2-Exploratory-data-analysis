# Exploratory data analysis

```python
# Identifying the top 10 most expensive countries based on the Cost of Living Index
top_10_expensive = df.nlargest(10, 'Cost of Living Index')[['Country', 'Cost of Living Index']]

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x='Cost of Living Index', y='Country', data=top_10_expensive, palette='viridis')
plt.title('Top 10 Most Expensive Countries by Cost of Living Index')
plt.xlabel('Cost of Living Index')
plt.ylabel('Country')

plt.tight_layout()
plt.show()
```

```python
# Identifying the top 10 cheapest countries based on the Cost of Living Index
top_10_cheap = df.nsmallest(10, 'Cost of Living Index')[['Country', 'Cost of Living Index']]

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x='Cost of Living Index', y='Country', data=top_10_cheap, palette='plasma')
plt.title('Top 10 Cheapest Countries by Cost of Living Index')
plt.xlabel('Cost of Living Index')
plt.ylabel('Country')

plt.tight_layout()
plt.show()
```

```python
# Extracting the relevant columns
relevant_columns = ['Country', 'Rent Index']
top_10_expensive = df.nlargest(10, 'Cost of Living Index')[relevant_columns]

# Plotting the Rent Index
plt.figure(figsize=(10, 6))
sns.barplot(x='Rent Index', y='Country', data=top_10_expensive, palette='viridis')
plt.title('Rent Index for Top 10 Most Expensive Countries')
plt.xlabel('Rent Index')
plt.ylabel('Country')

plt.tight_layout()
plt.show()
```

```python
# Extracting the relevant columns for Local Purchasing Power Index analysis
top_10_expensive_purchasing_power = df.nlargest(10, 'Cost of Living Index')[['Country', 'Local Purchasing Power Index']]

# Plotting the Local Purchasing Power Index
plt.figure(figsize=(10, 6))
sns.barplot(x='Local Purchasing Power Index', y='Country', data=top_10_expensive_purchasing_power, palette='cividis')
plt.title('Local Purchasing Power Index for Top 10 Most Expensive Countries')
plt.xlabel('Local Purchasing Power Index')
plt.ylabel('Country')

plt.tight_layout()
plt.show()
```

```python
# Setting the plot style
sns.set(style="whitegrid")

# Preparing the data in a suitable format for seaborn
df_melted = df.melt(value_vars=['Cost of Living Index', 'Rent Index', 'Groceries Index', 
                                'Restaurant Price Index', 'Local Purchasing Power Index'], 
                    var_name='Index', value_name='Value')

# Creating the combined box plot
plt.figure(figsize=(12, 8))
sns.boxplot(x='Index', y='Value', data=df_melted, palette='Set3')
plt.title('Comparison of Indices via Box Plots')
plt.xlabel('Index')
plt.ylabel('Value')
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```
